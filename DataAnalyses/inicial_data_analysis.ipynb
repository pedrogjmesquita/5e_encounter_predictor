```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../results_v2.csv', encoding='utf-8')
```

```python
df.info()
```

```python
df[f'p{1}_class'].value_counts()
```

```python
df.dificulty.mean()
```

```python
sns.histplot(df.dificulty, kde=True)
```

```python
df = df[df['p1_class'] != '0']
df = df.query('monster_cr != -1')
df = df[df['p1_class'] != '9608']
df = df[df['p1_level'] <= 5]
```

```python
df[f'p1_level'].value_counts()
```

```python
classes_ocurrency = []
for i in range(1,5):
    classes_ocurrency.append(df[f'p{i}_class'].tolist())
classes_ocurrency = classes_ocurrency[0]
```

```python
fig = plt.figure(figsize=(10, 7))
ax = sns.histplot(classes_ocurrency, kde=True)
ax.set_title('Ocorrência de classes')
ax.set_xlabel('Classes')
ax.set_ylabel('Ocorrência')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right');
```

```python
fig = plt.figure(figsize=(10, 7))
ax = sns.histplot(df.p1_level)
ax.set_title('Ocorrência de classes')
ax.set_xlabel('Classes')
ax.set_ylabel('Ocorrência')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right');
```

```python
lista = list((df.monster_cr.value_counts()).index)
lista.sort()
lista2=df.monster_cr.value_counts().sort_index(ascending=True).values.tolist()
fig = plt.figure(figsize=(10, 7))
ax = sns.barplot(x=lista, y=lista2)
ax.set_title('Ocorrência de CR')
ax.set_xlabel('CR')
ax.set_ylabel('Ocorrência')
```

```python
fig = plt.figure(figsize=(10, 7))
ax = sns.boxplot(x = df.dificulty)
```

```python
fig = plt.figure(figsize=(10, 7))
ax = sns.boxplot(x = df.monster_hp)
```

```python
fig = plt.figure(figsize=(10, 7))
ax = sns.boxplot(x = df.monster_ac)
```

```python
encontros_por_level = df.groupby('p1_level')
encontros_por_level.dificulty.mean()
```
