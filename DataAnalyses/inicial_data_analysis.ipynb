```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../results.csv', encoding='utf-8')
```

```python
df.info()
```

```python
df.final_total_hp.mean()
```

```python
sns.histplot(df.final_total_hp, kde=True)
```

```python
classes_ocurrency = []
for i in range(1,5):
    classes_ocurrency.append(df[f'p{i}_class'].tolist())
classes_ocurrency = classes_ocurrency[0]
```

```python
fig = plt.figure(figsize=(10, 7))
ax = sns.histplot(classes_ocurrency, kde=True)
ax.set_title('Ocorrência de classes')
ax.set_xlabel('Classes')
ax.set_ylabel('Ocorrência')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right');
```

```python
df = df.query('monster_cr != -1')
```

```python
fig = plt.figure(figsize=(10, 7))
ax = sns.histplot(df.monster_cr, bins=20, kde=True)
ax.set_title('Ocorrência de CR dos monstros')
ax.set_xlabel('CR')
ax.set_ylabel('Ocorrência')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right');
```
